## PAIN dataset

```python
def load_PAIN(db,sbj,f_bank,vwt,new_fs):

    channels_names = np.array(['Fp1','Fp2',
                      'F3','F4','C3','C4','P3','P4','O1','O2','F7','F8',
                      'T7','T8','P7','P8','Fz','Cz','Pz','Oz',
                      'FC1','FC2','CP1','CP2','FC5','FC6','CP5','CP6',
                      'TP9','TP10','LE','RE','P1','P2','C1','C2',
                      'FT9','FT10','AF3','AF4','FC3','FC4','CP3','CP4','PO3','PO4',
                      'F5','F6','C5','C6','P5','P6','PO9','Iz','FT7','FT8',
                      'TP7','TP8','PO7','PO8','Fpz','PO10','CPz','POz',
                      'Ne','Ma','Ext','ECG'])
    
    with open('{}BMOP_Motor_S{}.pkl'.format(db,'0' + str(sbj) if sbj < 10 else sbj), 'rb') as f:
        data = pickle.load(f)
        
    X = data['X']  # trials, channels, time
    y = data['y']
    sex = data['sex'].ravel()
    age = data['age'].ravel()
    fs = float(data['fs'])
    
    tf_repr = TimeFrequencyRpr(sfreq = fs, f_bank = f_bank, vwt = vwt)
    
    #Read electrode positions to load the best standard montage-MNE
    best_montages = get_best_montage(channels_names)
    montage = best_montages.iloc[0]['montage']
    no_channels = np.array(best_montages.iloc[0]['missings channels'])
    channels_to_remove = np.array([np.argwhere(channels_names==no)[0] for no in no_channels])[:,0]

    #Delete the missing channels the original array respecting the positions
    channels_names = np.delete(channels_names, channels_to_remove)
    X = np.delete(X, channels_to_remove, axis=1)

    #Number channels does not match with the dimension of X, 
    #thus the last channel is discarded because it has weird amplitudes
    X = X[:,:-1,:]

    info = mne.create_info(list(channels_names), sfreq=fs, ch_types="eeg")
    info.set_montage(montage)
    info

    event_id = {
        'pain/high':2,
        'resting':3,
        }

    events = [[i, 1, cls[0]] for i, cls in enumerate(y)]
    tmin = 0

    epochs = mne.EpochsArray(X, info, events=events, tmin=tmin, event_id=event_id)
    X = epochs.get_data()
    y = y-2
    X = np.squeeze(tf_repr.transform(X))
                             
    #Resampling
    if new_fs != fs:
        X = resample(X, int((X.shape[-1]/fs)*new_fs), axis = -1)
        
    return X,y,age,sex,fs,info
```

```python
def load_demo_info(db,sbj,f_bank,vwt,new_fs):
    with open('{}BMOP_Motor_S{}.pkl'.format(db,'0' + str(sbj) if sbj < 10 else sbj), 'rb') as f:
        data = pickle.load(f)

    sex = data['sex'].ravel()[0]
    age = data['age'].ravel()[0]
        
    return age,sex
```

```python
import matplotlib.pyplot as plt
import numpy as np

def makeStdBarPlot(data, target, title, model, xlabel, ylabel, width=10, height=8):
    colors_ = ['yellow','cyan','darkred','darkblue', 'khaki']
    
    x = []
    y = []
    colors = []
    stdy = []
    colors = []
    
    for dict_ in data:
        keys = list(dict_.keys())
        x += keys
        y += [dict_[key][target] for key in keys]
        stdy += [dict_[key]['Std' + target] for key in keys]
        colors += colors_[:len(keys)]
    
    plt.figure(figsize=(width, height))
    plt.bar(x, y, color=colors)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.errorbar(x, y, stdy, fmt='.', color='Black', elinewidth=2,capthick=10,errorevery=1, alpha=0.5, ms=4, capsize = 2)
    plt.ylim(0,1)
    plt.savefig(title + ".pdf", format="pdf")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size': 30})

def makeStdBarPlotCompatison(data, target, title, models, xlabel, ylabel, width=20, height=12):
    colors_ = ['yellow','cyan','darkred','darkblue', 'khaki']
    
    plt.figure(figsize=(width, height), dpi=120)
    for i, data_ in enumerate(data):
        x = []
        y = []
        colors = []
        stdy = []
        colors = []

        for dict_ in data_:
            keys = list(dict_.keys())
            x += keys
            y += [dict_[key][target] for key in keys]
            stdy += [dict_[key]['Std' + target] for key in keys]
            colors += colors_[:len(keys)]

        
        plt.subplot(1, 2, i + 1)
        plt.bar(x, y, color=colors)
        plt.title(title + ' - ' + models[i])
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.errorbar(x, y, stdy, fmt='.', color='Black', elinewidth=2,capthick=10,errorevery=1, alpha=0.5, ms=4, capsize = 2)
        plt.ylim(0,1)
        plt.xticks(rotation=40)
        plt.tight_layout()
    plt.savefig(title + "_Comp.png")
    plt.savefig(title + "_Comp.png", format="pdf")
    plt.show()
```

```python
accuraies_gfc = []
std_acc_gfc = []
kappas_gfc = []
std_kappa_gfc = []
auc_gfc = []
std_auc_gfc = []
accuracies_eegnet = []
std_acc_eegnet = []
kappas_eegnet = []
std_kappa_eegnet = []
auc_eegnet = []
std_auc_eegnet = []

for gfc, eegnet in zip([sDepGroupsResultsGfc, groupsResultsGfc, losoGroupsResultsGfc], [sDepGroupsResults, groupsResults, losoGroupsResults]):
    gfcAcc = np.mean([gfc[group]['Accuracy'] for group in list(gfc)])
    accuraies_gfc.append(gfcAcc)
    gfcAccStd = np.mean([gfc[group]['StdAccuracy'] for group in list(gfc)])
    std_acc_gfc.append(gfcAccStd)
    gfcKappa = np.mean([gfc[group]['Kappa'] for group in list(gfc)])
    kappas_gfc.append(gfcKappa)
    gfcKappaStd = np.mean([gfc[group]['StdKappa'] for group in list(gfc)])
    std_kappa_gfc.append(gfcKappaStd)
    gfcAuc = np.mean([gfc[group]['AUC'] for group in list(gfc)])
    auc_gfc.append(gfcAuc)
    gfcAucStd = np.mean([gfc[group]['StdAUC'] for group in list(gfc)])
    std_auc_gfc.append(gfcAucStd)
    
    
    eegnetAcc = np.mean([eegnet[group]['Accuracy'] for group in list(eegnet)])
    accuracies_eegnet.append(eegnetAcc)
    eegnetAccStd = np.mean([eegnet[group]['StdAccuracy'] for group in list(eegnet)])
    std_acc_eegnet.append(eegnetAccStd)
    eegnetKappa = np.mean([eegnet[group]['Kappa'] for group in list(eegnet)])
    kappas_eegnet.append(eegnetKappa)
    eegnetKappaStd = np.mean([eegnet[group]['StdKappa'] for group in list(eegnet)])
    std_kappa_eegnet.append(eegnetKappaStd)
    eegnetAuc = np.mean([eegnet[group]['AUC'] for group in list(eegnet)])
    auc_eegnet.append(eegnetAuc)
    eegnetAucStd = np.mean([eegnet[group]['StdAUC'] for group in list(eegnet)])
    std_auc_eegnet.append(eegnetAucStd)
    
print(f"\nGFC ACC: {np.mean(accuraies_gfc)} +- {np.mean(std_acc_gfc)}, EEGNet ACC: {np.mean(accuracies_eegnet)} +- {np.mean(std_acc_eegnet)}")
print(f"\nGFC Kappa: {np.mean(kappas_gfc)} +- {np.mean(std_kappa_gfc)}, EEGNet Kappa: {np.mean(kappas_eegnet)} +- {np.mean(std_kappa_eegnet)}")
print(f"\nGFC AUC: {np.mean(auc_gfc)} +- {np.mean(std_auc_gfc)}, EEGNet AUC: {np.mean(auc_eegnet)} +- {np.mean(std_auc_eegnet)}")
```
